```python
import pandas as pd
import matplotlib.pyplot as plt


# Group the data by job titles and remote work options
grouped_by_job_titles = df.groupby(['title', 'mode']).size().unstack()

# Group the data by companies and remote work options
grouped_by_companies = df.groupby(['company', 'mode']).size().unstack()

# Group the data by locations and remote work options
grouped_by_locations = df.groupby(['country', 'mode']).size().unstack()





```

```python
# Plot the distribution of remote work options by job titles

plt.figure(figsize=(12, 100)) 
grouped_by_job_titles.plot(kind='bar', stacked=True)
plt.xlabel('Job Titles')
plt.ylabel('Count')
plt.xticks(rotation = 90)
plt.tight_layout(pad=2)
plt.title('Distribution of Remote Work Options by Job Titles')
plt.legend(title='Remote Work Options')
plt.show()
```

```python

# Plot the distribution of remote work options by companies
grouped_by_companies.plot(kind='bar', stacked=True)
plt.xlabel('Companies')
plt.ylabel('Count')
plt.title('Distribution of Remote Work Options by Companies')
plt.legend(title='Remote Work Options')
plt.show()
```

```python
# Plot the distribution of remote work options by locations
grouped_by_locations.plot(kind='bar', stacked=True)
plt.xlabel('country')
plt.ylabel('Count')
plt.title('Distribution of Remote Work Options by Locations')
plt.legend(title='Remote Work Options')
plt.show()
```

```python
 #Cross-tabulate remote work options with criteria
cross_tab = pd.crosstab(df['mode'], df['criteria'])

# Calculate the percentage of each criteria for each remote work option
percentage_cross_tab = cross_tab.apply(lambda x: x / x.sum() * 100, axis=1)

# Plot the percentage distribution of criteria by remote work options
percentage_cross_tab.plot(kind='bar', stacked=True)
plt.xlabel('Remote Work Options')
plt.ylabel('Percentage')
plt.title('Distribution of Criteria by Remote Work Options')
plt.legend(title='Criteria')

plt.show()
```

# Temporal Analysis:

```python

# Group job postings by month and count the frequency
job_postings_by_month = df.groupby('month').size().reset_index(name='Job Postings')


# Plot the distribution of job postings over time
plt.figure(figsize=(12, 6))
plt.plot(job_postings_by_month['month'], job_postings_by_month['Job Postings'], marker='o')
plt.xlabel('Month')
plt.ylabel('Number of Job Postings')
plt.title('Distribution of Job Postings Over Time')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.grid(True)
plt.show()
```

Investigating the relationship between the posted date and other factors such as job titles.

```python
# Get the top 20 job titles
top_20_titles = df['title'].value_counts().head(20).index

# Filter the data for the top 20 titles
top_20_data = df[df['title'].isin(top_20_titles)]

# Convert 'posted_date' column to datetime format
top_20_data['posted_date'] = pd.to_datetime(top_20_data['posted_date'])

# Group job postings by date and title, count the frequency, and reshape the data
job_postings_by_date_title = top_20_data.groupby(['posted_date', 'title']).size().unstack().reset_index()

# Sort the dates in ascending order
job_postings_by_date_title = job_postings_by_date_title.sort_values('posted_date')

# Plot the distribution of job postings over time for each title
plt.figure(figsize=(12, 6))
for title in top_20_titles:
    plt.plot(job_postings_by_date_title['posted_date'], job_postings_by_date_title[title], label=title)
    # Annotate the title outside the graph
    last_value = job_postings_by_date_title[title].iloc[-1]
    plt.annotate(title, (job_postings_by_date_title['posted_date'].iloc[-1], last_value))

plt.xlabel('Date')
plt.ylabel('Number of Job Postings')
plt.title('Distribution of Job Postings Over Time for Top 20 Job Titles')
plt.legend()
plt.xticks(rotation=90)
plt.grid(True)
plt.tight_layout()
plt.show()
```
